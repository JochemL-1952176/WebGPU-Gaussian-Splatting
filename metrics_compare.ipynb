```python
import os
import pandas as pd
import seaborn as sns
```

```python
path = 'benchmarks/bicycle'
```

```python
methodNames = os.listdir(path)
methods = []
for methodName in methodNames:
	method = pd.read_pickle(path + '/' + methodName + '/data.pkl')
	method['method'] = methodName
	methods.append(method)

data = pd.concat(methods)
```

```python
sns.jointplot(data, x='SSIM', y='median frametime', hue='method')
sns.jointplot(data, x='PSNR', y='median frametime', hue='method')
```
